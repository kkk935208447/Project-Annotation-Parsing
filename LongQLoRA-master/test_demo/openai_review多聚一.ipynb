```python
import requests as requests
```

```python
# 在这里配置您在本站的API_KEY
api_key = "you api"
headers = {
    "Authorization": 'Bearer ' + api_key,
}
```

```python
question = INPUT_PROMPT.replace('{{Reviews}}',input)
print(SYSTEM_PROMPT)
print("-"*100)
print(question)
```

```python
params = {
    "messages": 
    [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": question}
        ],
    # 如果需要切换模型，在这里修改
    "model": 'claude-3-sonnet-20240229',
    # 输出强制为 json格式
    "response_format":{ "type": "json_object" }
}
response = requests.post(
    "XXXXXXX/v1/chat/completions",  # XXXX是三方站 url
    headers=headers,
    json=params,
    stream=False
)
res = response.json()
res_content = res["choices"][0]['message']['content']
print(res_content)
```
